```python
#Q1 >> Write a Python function that takes a list of numbers as input and returns 
#      the sum of all even numbers in the list.

def even_num(numbers):
    return sum(num for num in numbers if num%2 == 0 )
even_num([1,2,3,4,5])
    
```

```python
#Q3 >> Implement a Python function that takes a list of integers and returns a 
#     new list containing the squares of each number.

def squar_list(numbers):
    return [num**2 for num in numbers]
    
squar_list([1,2,3,5])
```

```python
#Q4 >> Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [n for n in range(1, 201) if is_prime(n)]

print(prime_numbers)
```

```python
#Q6 >> Write a generator function in Python that yields the powers of 2 up to a given exponent.

def square_number(n):
    for i in range(n):
        yield (i**2)  
list(square_number(10))
```

```python
#Q7 >> Implement a generator function that reads a file line by line and yields each line as a string.
def read_string_line_by_line(input_string):
    for line in input_string.splitlines(" "):
        yield line
    
test_string = """I
am
DS 
student
"""
for line in read_string_line_by_line(test_string):
        print(line)
```

```python
#Q8 >> Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

tuples_list = [(1, 3), (4, 1), (2, 2), (5, 0)]

sorted_list = sorted(tuples_list, key=lambda x: x[1])

print(sorted_list)

```

```python
#Q9 >> Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius_to_fah = lambda c: (c * 9/5) + 32
celsius_temps = [0, 20, 37, 100]
fah_temps = list(map(celsius_to_fah, celsius_temps))

print(fah_temps)

```

```python
#Q10 >> Create a Python program that uses `filter()` to remove all the vowels from a given string.


not_vowel = lambda x: x not in "aeiou"
def remove_vowels(sentance):
    return "".join(filter(not_vowel,sentance))

text = "Pwskills"
result = remove_vowels(text)
print(result)
```
